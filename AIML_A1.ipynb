```python
# 1 - Implement Basic Data Structures using Numpy, Pandas
import numpy as np
import pandas as pd

# 1. Lists
list1 = [1, 2, 3, 4]
list2 = [5, 6, 7, 8]
print(type(list1))
```

```python
#arrays
array1 = np.array(list1)
array2 = np.array(list2)
print(type(array1))
```

```python
#4.Mathematical calculations
#5.combining text

mult = array1*array2
print('multiplication =',mult)

div = array2/array1
print('division=',div)

power = array1**array2
print('poweroff=',power)
```

```python
#implementation of sin log log2 exp
sin = np.sin(array1)
print(sin)
log = np.log(array1)
print(log)
log2 = np.log2(array1)
print(log2)
exp = np.exp(array1)
print(exp)
```

```python
# Part 2 - Visualization of Data using matplotlib, pyplots Packages
import matplotlib.pyplot as plt

# Generate a Data set for Health Care
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

```python
plt.scatter(age,height,label = 'height vs age')
plt.xlabel('age')
plt.ylabel('height')
plt.title('scatter plot - height vs age')
plt.legend()
plt.show()
```

```python
# 4. Bar Chart
plt.bar(people, weight, label='Weight')
plt.xlabel('People')
plt.ylabel('Weight')
plt.title('Bar Chart - Weight of People')
plt.legend()
plt.show()
```

```python
# 5. Histogram
plt.hist(age, bins=5, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram - Age Distribution')
plt.show()
```
